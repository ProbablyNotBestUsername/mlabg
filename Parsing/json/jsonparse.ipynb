```python
import pandas as pd
import json
import glob
import re
from numpy import float32

import warnings
warnings.filterwarnings('ignore')
```

```python
all_json = glob.glob(r".\\files\*.json")
```

```python
def open_json(filename):
    file = open(filename, encoding='utf-8')
    return json.load(file)
```

```python
with open(".\\files\Cybersport.Metaratings.json", encoding='utf-8') as f:
    print(json.load(f))
```

```python
ArticleText = []
PublishDate = []
Rating = []

for file in all_json:
    json_decoded = open_json(file)
    if json_decoded['refs']:
        print(f'file {file}')
        if json_decoded['refs'][0]:
            article_text = json_decoded['refs'][0][0]

            Date = json_decoded['refs'][0][1]
            day = Date['day']
            month = Date['month']
            time = Date['time']
            publish_date = f'{day} {month} в {time}'

            rating = json_decoded['refs'][0][2]['rate']

            ArticleText.append(article_text)
            PublishDate.append(publish_date)
            Rating.append(rating)
            
```

```python
df = pd.DataFrame({
    'Author': pd.Series(dtype=str),
    'PublishDate': pd.Series(dtype=str),
    'Rating': pd.Series(dtype=float32),
    'ArticleText': pd.Series(dtype=str),
    'CompanyName': pd.Series(dtype=str),
    'CompanyDescription': pd.Series(dtype=str)
})
```

```python
df['PublishDate'] = PublishDate
df['Rating'] = Rating
df['ArticleText'] = ArticleText
```
